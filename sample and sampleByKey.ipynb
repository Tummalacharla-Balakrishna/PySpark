```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Sample').getOrCreate()
```

```python
rdd = spark.sparkContext.parallelize(range(1000))

# to generate random numbers 
rdd.sample(False,0.1,10).collect()
```

```python
dict = {'a':0.01,'b':0.04}
rdd2 = spark.sparkContext.parallelize(dict.keys()).cartesian(spark.sparkContext.parallelize(range(0,1000)))
rdd2.sampleByKey(False,dict).collect()
```
